```python
print("Простой калькулятор (+ / - *). Пустой ввод на любом шаге - выход")
## Разделитель пробелами
sep = "\u2009"

while True:
    ## ВВод a
    a_str = input("a: ")
    if a_str == "":
        break
    else:
        a = float(a_str)
        
    ## ВВод op
    op = input("op: ")
    if op == "":
        break
        
    ## Проверка оператора op        
    if op not in ["+", "-", "/", "*"]:
        print("Ошибка: неизвестный оператор")
        continue
    
     ## ВВод b       
    b_str = input("b: ")
    if b_str == "":
        break
    else:
        b = float(b_str)

    ## Проверка деления на ноль
    if op == "/" and b == 0:
        print("Ошибка: деление на ноль")
        continue

    ## Этап 1: Вычисление
    if op == "+":
        value = a + b
    elif op == "-":
        value = a - b
    elif op == "/":
        value = a / b
    elif op == "*":
        value = a * b
    else:
        print("Ошибка: неизвестный оператор")
        continue

    ## Этап 2: Форматирование
    if value % 1 == 0:
        ## если целое число
        result = f"{value:,.0f}"
    else:
        ## если дробное число
        result = f"{value:,.5f}"
        result = result.rstrip("0").rstrip(".")

    ## Замена запятых на пробелы
    result = result.replace(",", sep)
    
    print("= ", result)
 
```
